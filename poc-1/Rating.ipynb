1.1 Read movies and merge rantings

```python
import numpy as np
import pandas as pd

movies = pd.read_csv('Movies.csv', sep=',')
ratings = pd.read_csv('Ratings.csv', sep=';')

df = movies.merge(ratings, on="movieId")

df.head()
```

2.1 Filter data

```python
movies_table = pd.pivot_table(df, index='title', columns='userId', values='rating').fillna(0)
movies_table.head()
```

```python
import sklearn.metrics.pairwise as pw

rec = pw.cosine_similarity(movies_table)
rec_df = pd.DataFrame(rec, columns=movies_table.index, index=movies_table.index)
rec_df.head()
```

```python
cossine_df = pd.DataFrame(rec_df['Thor'].sort_values(ascending=False))
cossine_df.columns = ['Recommendations']
cossine_df.head(20)
```
